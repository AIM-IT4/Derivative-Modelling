```python
import numpy as np
import matplotlib.pyplot as plt

def constant_intensity_cds(n, R, spread, recovery):
    """Valuation of a credit default swap under constant intensity model.
    
    Args:
        n (int): Number of periods.
        R (float): Risk-free interest rate.
        spread (float): Spread of the credit default swap.
        recovery (float): Recovery rate.
        
    Returns:
        float: Present value of the credit default swap.
    """
    dt = 1 / n
    q = spread * dt
    df = np.exp(-R * np.arange(1, n + 1) * dt)
    
    pv = 0.0
    for t in range(n):
        pv += q * df[t]
    
    pv *= recovery
    
    return pv

def plot_cds_valuation(n_values, pv_values, model_names):
    """Plots the valuation of credit default swaps using different models.
    
    Args:
        n_values (list): List of periods.
        pv_values (list): List of present values for each model.
        model_names (list): List of model names.
    """
    plt.figure(figsize=(8, 6))
    for i in range(len(model_names)):
        plt.plot(n_values, pv_values[i], label=model_names[i])
    
    plt.xlabel('Number of Periods')
    plt.ylabel('Present Value')
    plt.title('Valuation of Credit Default Swaps')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
n_values = [1, 2, 3, 4, 5]
spread = 0.02
recovery = 0.4
R = 0.05

model_names = ['Constant Intensity']
pv_values = []

for n in n_values:
    pv = constant_intensity_cds(n, R, spread, recovery)
    pv_values.append(pv)

plot_cds_valuation(n_values, [pv_values], model_names)
```
